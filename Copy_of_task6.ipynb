```python
from matplotlib import pyplot as plt
days=[1,2,3,4,5,6,7,8]
max_speed=[80,91,92,88,77,79,76,75]
plt.plot(days,max_speed,'r',label='line 1',linewidth=6)
plt.legend()
plt.title('car speed measurement')
plt.xlabel('days')
plt.ylabel('max_speed')
plt.grid()
plt.show()


days=[1,2,3,4,5,6,7,8]
min_speed=[42,43,40,42,33,36,34,35]
plt.plot(days,min_speed,'g',label='line 2',linewidth=5)
plt.legend()
plt.title('car speed measurement')
plt.xlabel('days')
plt.ylabel('min_speed')
plt.grid()
plt.show()


days=[1,2,3,4,5,6,7,8]
avg_speed=[46,58,57,56,40,42,41,36]
plt.plot(days,avg_speed,'o-',label='line 3',linewidth=4)
plt.title('car speed measurement')
plt.xlabel('days')
plt.ylabel('avg_speed')
plt.grid()
plt.show()


```

```python
from matplotlib import pyplot as plt
languages=['python','SQL','java','c++','javascript']
plt.bar(languages,[56,39,34,34,29],label='popularity',color='g',width=.3)
plt.legend()
plt.title('popularity')
plt.show()
plt.bar(languages,[44,36,55,50,42],label='security',color='r',width=.4)
plt.legend()
plt.title('security')
plt.show()
languages=['python','SQL','java','c++','javascript']
plt.barh(languages,[56,39,34,34,29],height=.2,label='popularity',color='b')
plt.legend()
plt.title('popularity')
plt.show()
plt.barh(languages,[44,36,55,50,42],height=.5,label='security')
plt.legend()
plt.title('security')
plt.show()

```
